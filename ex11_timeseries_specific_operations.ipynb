````python
import yfinance as yf
import pandas as pd
from matplotlib import pyplot as plt
````

````python
# 구글 주식 데이터 로딩
goog = yf.download("GOOG", start="2004-01-01", end="2024-12-31") #기본적으로 일별 데이터 가져온다.
goog.head()
````

````python
# [+] 종가(Close) 데이터만 사용하기
goog = goog['Close']
goog
````

````python
# [+] Lineplot 시각화
goog.plot()
````

#### <strong> 리샘플링(Resampling)
+ ```resample()```: 빈도를 기준으로 집계 연산
+ ```asfreq()```: 빈도 단위 별 데이터 선택

````python
# [+] 다운샘플링: resample() resample은 집계하고 평균
goog_downsampled = goog.resample('ME').mean() #월별 말일 기준으로 집계 뒤에 .mean 평균값 구하는 메소드 
goog_downsampled #일별데이터에서 월별데이터로 다운샘플링 했다.
````

````python
#다운샘플링 시각화
fig, ax = plt.subplots(2, 1, sharey=True, figsize=(12, 6))
goog.plot(label='original', ax=ax[0])
ax[0].legend()
goog_downsampled.plot(label='downsampled', ax=ax[1])
ax[1].legend()
plt.tight_layout()
````

````python
# [+] 다운샘플링: asfreq() 해당되는 데이터만 가져온다.
goog_downsampled = goog.asfreq('ME') #중간중간에 null값 
print(goog_downsampled)

goog_downsampled.plot()
````

````python
# [+] 업샘플링: resample()
goog_upsampled = goog.resample('12h').mean() 
print(goog_upsampled) #없는 데이터는 NaN값 
````

#### <strong> 시프팅(Shifting)
+ ```shift(periods)``` : ```periods``` 값 만큼 시계열 이동

````python
# [+] 900일 뒤로 시프팅 적용
goog_shifted = goog.shift(periods=900)
goog_shifted
````

````python

fig, ax = plt.subplots(2, 1, sharey=True, figsize=(12, 6))

# Legends and annotations
local_max = pd.to_datetime('2007-11-05')
offset = pd.Timedelta(900, 'D')

goog.plot(label='original', ax=ax[0])
ax[0].legend()
ax[0].get_xticklabels()[2].set(weight='heavy', color='red')
ax[0].axvline(local_max, alpha=0.3, color='red')

goog_shifted.plot(label='shifted 900 days', ax=ax[1])
ax[1].legend()
ax[1].get_xticklabels()[2].set(weight='heavy', color='red')
ax[1].axvline(local_max + offset, alpha=0.3, color='red')

plt.tight_layout()
````

````python
# shifting 응용: 주가 변화율 계산
goog_shifted = goog.shift(periods=1) # [+] 시프팅
(goog - goog-goog_shifted) / goog_shifted # 일부 주가 변화율 계산
````

````python
# shifting 응용: 차분 시계열 구하기
goog_diff = goog - goog_shifted  # [+] 차분
print(goog_diff)

goog_diff.plot()
````

#### <strong> 윈도잉(Windowing)

````python
# [+] 365일 단위 윈도잉
goog_rolled = goog.rolling(365) 
print(goog_rolled)

goog_rolled.mean().plot()
````

````python
data = pd.DataFrame({'input': goog.values.flatten(),
                     # moving average
                     'one-year rolling_mean': goog_rolled.mean().values.flatten(), 
                     # standard deviation
                     'one-year rolling_std': goog_rolled.std().values.flatten()}, index=goog.index)
ax = data.plot(style=['-', '--', ':'])
ax.lines[0].set_alpha(0.3)
````
